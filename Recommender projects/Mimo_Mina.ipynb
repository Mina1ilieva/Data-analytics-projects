```python
#let's make each key a column so I could use these for my analyses.
key_values = {}
for cell in data_test["labelling"]:
    for key, value in cell.items():
        if key not in key_values:
            key_values[key] = set()
        if isinstance(value, list):
            for elem in value:
                key_values[key].add(elem)
        else:
            key_values[key].add(value)
for key, values in key_values.items():
    print(f"Key: {key}, Values: {', '.join(map(str, values))}")
```

```python
for key in key_values:
    def get_value(cell, key):
        if key in cell:
            value = cell[key]
            if isinstance(value, list):
                return ', '.join(map(str, value))  # convert list to string
            else:
                return value
        else:
            return None  

    data_test[key] = data_test["labelling"].apply(lambda x: get_value(x, key))
```

```python
#Let's prepare the hashtags for vectorization: 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
#convert all hashtags to strings 
data_test_copy['hashtags'] = data_test_copy['hashtags'].astype(str)
tokenizer = Tokenizer()
tokenizer.fit_on_texts(data_test_copy['hashtags'])
sequences = tokenizer.texts_to_sequences(data_test_copy['hashtags']) #convert them to sequences 
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')
for i in range(max_length):
    data_test_copy[f'hashtag_{i+1}'] = padded_sequences[:, i]
#print(data_test_copy['hashtags'].unique())
data_test_copy.isnull()
```

```python
#let's do the same for comments
data_test_copy['comments'] = data_test_copy['comments'].astype(str)
tokenizer_comments = Tokenizer()
tokenizer_comments.fit_on_texts(data_test_copy['comments'])
sequences_comments = tokenizer_comments.texts_to_sequences(data_test_copy['comments'])
max_length_comments = max(len(seq) for seq in sequences_comments)
padded_comments = pad_sequences(sequences_comments, maxlen=max_length_comments, padding='post')
for i in range(max_length_comments):
    data_test_copy[f'comment_{i+1}'] = padded_comments[:, i]
data_test_copy.isnull()
```
